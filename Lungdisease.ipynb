```python
data=data.drop('Patient Id',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
data['Level'] = encoder.fit_transform(data['Level'])
```

```python
X=data.drop('Level',axis=1)
y=data['Level']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
!pip install xgboost
```

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
params = {
    'objective': 'multi:softmax',
    'num_class': 3,
    'max_depth': 1,             # Reduce tree depth for simpler trees
    'eta': 0.03,                # Lower learning rate for gradual learning
    'lambda': 15,               # Increase L2 regularization
    'alpha': 7,                 # Increase L1 regularization
    'subsample': 0.7,           # Use 70% of the data per tree
    'colsample_bytree': 0.7,    # Use 70% of features per tree
    'min_child_weight': 4,      # Prevent small leaf nodes
    'eval_metric': 'merror'
}
```

```python
# Train the XGBoost model
num_rounds = 70
model = xgb.train(params, dtrain, num_rounds)
```

```python
# Make predictions on the test set
y_pred = model.predict(dtest)
```

```python
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Save the trained model to a file
import joblib

# Replace 'model.joblib' with your desired filename
joblib.dump(model, 'lung_disease_model.joblib')

print("Model saved successfully as 'lung_disease_model.joblib'")
```
